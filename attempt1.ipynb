```python
import requests

def search_newspaper_titles(query, format='json', page=1):
    url = f"https://chroniclingamerica.loc.gov/search/titles/results/?terms={query}&format={format}&page={page}"
    response = requests.get(url)
    return response.json() if format == 'json' else response.text

def search_newspaper_pages(query, format='json', page=1):
    url = f"https://chroniclingamerica.loc.gov/search/pages/results/?andtext={query}&format={format}&page={page}"
    response = requests.get(url)
    return response.json() if format == 'json' else response.text

def get_json_view(resource):
    url = f"https://chroniclingamerica.loc.gov/{resource}.json"
    response = requests.get(url)
    return response.json()

def get_linked_data(resource):
    url = f"https://chroniclingamerica.loc.gov/{resource}.rdf"
    response = requests.get(url)
    return response.text

if __name__ == "__main__":
    query = "Pennsylvania"
    print("Searching newspaper titles...")
    titles = search_newspaper_titles(query)
    print(titles)

    print("Searching newspaper pages...")
    pages = search_newspaper_pages(query)
    print(pages)

    print("Fetching JSON view of a newspaper title...")
    json_data = get_json_view("lccn/sn86069873")
    print(json_data)

    print("Fetching linked data view of a newspaper title...")
    rdf_data = get_linked_data("lccn/sn85038615")
    print(rdf_data)
```
